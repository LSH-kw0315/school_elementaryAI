```python
df_list=[]
for i in range(2010,2021):
    for j in range(1,13):
        year=str(i)
        month=str(j).zfill(2)
        df=None
        try:
            df=pd.read_excel('kto_{}.xlsx'.format(year+month),header=1,usecols="A:G",skipfooter=4)
        except:
            continue
        df['(%)']=df[""]/df[""]*100
        df['(%)']=df[""]/df[""].sum()*100
        df[""]=year
        df[""]=month
        df[""]="{}-{}".format(year,month)
        df_list.append(df)
    
```

```python
len(df_list)
```

```python
df=pd.concat(df_list,ignore_index=True)
```

```python
df[df['']=='']
```

```python
for i in df[''].unique():
    target=df[df['']==i]
    target.to_excel("[ ]-{}.xlsx".format(i))
```
