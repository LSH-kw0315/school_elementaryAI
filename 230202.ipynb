```python
import matplotlib.pyplot as plt
import numpy as np
import random
from tensorflow import keras
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import load_iris
from sklearn.datasets import load_boston
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
```

```python
x1=np.array([2,4,6,8])
x2=np.array([1,2,1,3])
y=np.array([20,75,55,85])

learning_rate=0.00002
a1=0
a2=0
b=0

for i in range(10000):
    y_hat=a1*x1+a2*x2+b
    error=y_hat-y
    
    a1_diff=sum(2*error*x1)
    a2_diff=sum(2*error*x2)
    b_diff=sum(2*error)

    a1=a1-a1_diff*learning_rate
    a2=a2-a2_diff*learning_rate
    b=b-b_diff*learning_rate

    print("{}회 학습 // 기울기1 : {} // 기울기2 : {} // 절편 : {}".format(i+1,a1,a2,b))
```

```python
x1=np.arange(-3,3,0.1)
y=np.sin(x1)

learning_rate=0.00001
a=random.random()
b=random.random()
c=random.random()
d=random.random()
e=random.random()
for i in range(20000):
    y_hat=a*x1**4+b*x1**3+c*x1**2+d*x1+e
    error=y_hat-y

    a_diff=sum(2*error*x1**4)
    b_diff=sum(2*error*x1**3)
    c_diff=sum(2*error*x1**2)
    d_diff=sum(2*error*x1)
    e_diff=sum(2*error)

    a=a-a_diff*learning_rate
    b=b-b_diff*learning_rate
    c=c-c_diff*learning_rate
    d=d-d_diff*learning_rate
    e=e-e_diff*learning_rate

    print("{}회 학습 // a : {} // b : {} // c : {} // d:{} // e:{}".format(i+1,a,b,c,d,e))
```

```python
iris=load_iris()

data, target = iris['data'],iris['target']

train_input, test_input, train_target, test_target = train_test_split(data,target,test_size=0.2)


ss=StandardScaler()
train_input_scaled=ss.fit_transform(train_input)
test_input_scaled=ss.transform(test_input)

train_target_categorized=keras.utils.to_categorical(train_target)
test_target_categorized=keras.utils.to_categorical(test_target)

model=keras.Sequential()
model.add(keras.layers.Dense(8,activation='relu',input_shape=(4,)))
model.add(keras.layers.Dense(3,activation='softmax'))

es=keras.callbacks.EarlyStopping(patience=15,restore_best_weights=True)
model.compile(loss='categorical_crossentropy',metrics='accuracy')
hist=model.fit(train_input_scaled,train_target_categorized,epochs=1000,validation_data=(test_input_scaled,test_target_categorized),callbacks=[es])
```

```python
plt.plot(hist.history['loss'])
plt.plot(hist.history['val_loss'])
plt.legend(['loss','val_loss'])
```
